```python
from ocean import *
from matplotlib import pyplot as plot
%matplotlib inline
```

```python
ocean = Ocean(100, 100)
pictures_number = 3
picture_duration = 5000
```

```python
for i in range(pictures_number):
    victims = []
    predators = []

    for i in range(picture_duration):
        victims.append(ocean.type_counter(Victim))
        predators.append(ocean.type_counter(Predator))
        ocean.make_turn()

    plot.figure()
    plot.plot(victims, color='blue', label='victims')
    plot.plot(predators, color='red', label='predators')
    plot.gca().legend()
```
